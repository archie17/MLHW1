```python
#QUESTION-3-average price of BMW cars

df=pd.DataFrame(data)
df
df[df['Make']=="BMW"].MSRP.mean()
```

```python
#QUESTION-4-missing values in Engine HP after 2015
df1=df[df['Year']>=2015]
df1["Engine HP"].isnull().sum()
```

```python
#QUESTION-5 - average Engine HP,using mean and fillna
mean_hp_before=df1["Engine HP"].mean()
print(round(mean_hp_before))


```

```python
mean_hp_after=df1["Engine HP"].fillna(mean_hp_before).mean()
print(round(mean_hp_after))
```

```python
#QUESTION-6
df3=df[["Make","Engine HP","Engine Cylinders","highway MPG"]]
df3
```

```python
df3=df3[df3["Make"]=="Rolls-Royce"]
df3
```

```python
df3.drop(columns='Make',inplace=True)
df3
```

```python
df3=df3.drop_duplicates()
```

```python
df3
```

```python
X=df3.to_numpy()
print(X)
print("dimension:",X.ndim)
print("shape:",X.shape)
```

```python
X=np.asmatrix(X)
X
```

```python
XTX=(X.T).dot(X)
XTX
```

```python
XTXinv=np.linalg.inv(XTX)
print(XTXinv)
```
